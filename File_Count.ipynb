```python
x = []
for fols in os.listdir(base_dir):
    folder = os.path.join(base_dir,fols)
    for files  in os.listdir(folder):
        file = os.path.join(folder,files)
        if not os.path.getsize(file):
            print("Invalid file found which is present in: " + os.path.dirname(file) + '\ ' + '\033[1m' + os.path.basename(file) + '\033[0m')
            x.append(file)
        else:
            with open(file,'r') as b:
                line = b.readlines()
                for l in line:
                    if l.endswith('0 0 0 0'):
                        print('\033[0m' + "No bounding box detected in: " + os.path.dirname(file) + '\ ' + os.path.basename(file))
                        x.append(file)
print(len(x))
```

```python
# x = []
for fols in os.listdir(base_dir):
    folder = os.path.join(base_dir,fols)
    for files  in os.listdir(folder):
        file = os.path.join(folder,files)
        if file in x:
            os.remove(file)
```

# Rename

```python
some_dir = r'./test'
some_dir1 = r'./test1'
for files in os.listdir(some_dir):
    f = os.path.basename(files)
    file = f.split('.')
    for ano in os.listdir(some_dir1):
        annot = os.path.basename(ano)
        an = annot.split('.')
        if file[0] == an[0]:
            os.rename(os.path.join(some_dir,f), os.path.join(some_dir, file[0]+'s.jpg'))
            os.rename(os.path.join(some_dir1,annot), os.path.join(some_dir1, an[0]+'s.txt'))
```

# Final data preparation and checking

```python
import shutil

img_dir = r"./../WILO/Img"
ano_dir = r"./../WILO/Ann"
some_dir = r'./image'
some_dir1 = r'./annotation'

for fold in os.listdir(img_dir):
    folder = os.path.join(img_dir, fold)
    for files in os.listdir(folder):
        file = os.path.join(folder, files)
        shutil.copy(file, some_dir)
        
for fold in os.listdir(ano_dir):
    folder = os.path.join(ano_dir, fold)
    for files in os.listdir(folder):
        file = os.path.join(folder, files)
        shutil.copy(file, some_dir1)
```
